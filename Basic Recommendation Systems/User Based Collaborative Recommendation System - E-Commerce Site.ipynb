```python
import pandas as pd

df = pd.read_csv("../Data/ecommerce_data.csv", encoding="ISO-8859-1")
df.head()
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df.describe()
```

```python
df[df['CustomerID'].isna()].head()
```

```python
df = df.dropna(subset=['CustomerID'])
df.head()
```

```python
customer_item_matrix = df.pivot_table(index='CustomerID', columns='StockCode', values='Quantity')
customer_item_matrix.head()
```

```python
customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x > 0 else 0)
customer_item_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_user_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix))
user_user_sim_matrix.head()
```

```python
user_user_sim_matrix.columns = customer_item_matrix.index
user_user_sim_matrix['CustomerID'] = customer_item_matrix.index
user_user_sim_matrix = user_user_sim_matrix.set_index('CustomerID')
user_user_sim_matrix.head()
```

```python
def recommend_items_to_a_customer(customer):
    most_similar_customer = user_user_sim_matrix.loc[customer].sort_values(ascending=False).to_frame().index[1]
    items_customer_bought = []
    items_sim_customer_bought = []
    
    for i in range(customer_item_matrix.shape[1]):
        if (customer_item_matrix.loc[customer][i] == 1):
            items_customer_bought.append(customer_item_matrix.loc[customer].index[i])
        if (customer_item_matrix.loc[most_similar_customer][i] == 1):
            items_sim_customer_bought.append(customer_item_matrix.loc[most_similar_customer].index[i])
            
    recommend_items = set(items_sim_customer_bought) - set(items_customer_bought)
    return df.loc[df['StockCode'].isin(recommend_items), 
                  ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
```

```python
recommend_items_to_a_customer(12358)
```

```python
def recommend_ordered_items_to_a_customer(customer, n_similar_cust=5, n_items=2):
    most_similar_customers = user_user_sim_matrix.loc[customer].sort_values(ascending=False).to_frame().index[1:(n_similar_cust + 1)]
    items_customer_bought = []
    sim_customers_bought = []
    
    for cust in most_similar_customers:
        items_sim_customer_bought = []
        for i in range(customer_item_matrix.shape[1]):
            if (customer_item_matrix.loc[cust][i] == 1):
                items_sim_customer_bought.append(customer_item_matrix.loc[cust].index[i])
        sim_customers_bought.append(items_sim_customer_bought)
    
    for i in range(customer_item_matrix.shape[1]):
            if (customer_item_matrix.loc[customer][i] == 1):
                items_customer_bought.append(customer_item_matrix.loc[customer].index[i])
    
    recommend = dict()
    for i in range(n_similar_cust):
        diff = set(sim_customers_bought[i]) - set(items_customer_bought)
        for j in diff:
            if j in recommend:
                recommend[j] += 1
            else:
                recommend[j] = 1
        
    recommend = {k: v for k, v in sorted(recommend.items(), key=lambda item: item[1], reverse=True)}
    recommend = list(recommend)
    return df.loc[df['StockCode'].isin(recommend[:n_items]), 
                  ['StockCode', 'Description']].drop_duplicates().set_index('StockCode')
```
