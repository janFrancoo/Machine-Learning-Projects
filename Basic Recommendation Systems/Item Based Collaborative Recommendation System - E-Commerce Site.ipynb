```python
# This notebook is a continuation of the previous notebook, some steps are skipped

import pandas as pd

df = pd.read_csv("../Data/ecommerce_data.csv", encoding="ISO-8859-1")
df.head()
```

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df = df.dropna(subset=['CustomerID'])
```

```python
item_customer_matrix = df.pivot_table(index='StockCode', columns='CustomerID', values='Quantity')
item_customer_matrix.head()
```

```python
item_customer_matrix = item_customer_matrix.applymap(lambda x: 1 if x > 0 else 0)
item_customer_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_item_sim_matrix = pd.DataFrame(cosine_similarity(item_customer_matrix))
item_item_sim_matrix.head()
```

```python
item_item_sim_matrix.columns = item_customer_matrix.index
item_item_sim_matrix['StockCode'] = item_customer_matrix.index
item_item_sim_matrix = item_item_sim_matrix.set_index('StockCode')
item_item_sim_matrix.head()
```

```python
item_item_sim_matrix.loc["10080"].sort_values(ascending=False).to_frame()[1:11]
```

```python
def get_similar_items(item, n_recommend):
    most_similar_items = item_item_sim_matrix.loc[item].sort_values(ascending=False).to_frame().index[1:(n_recommend + 1)]            
    return df.loc[df['StockCode'].isin(most_similar_items), 
                  ['StockCode', 'Description']].drop_duplicates().set_index('StockCode').loc[most_similar_items]
```
