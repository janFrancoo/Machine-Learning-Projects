```python
import pandas as pd

labs_df = pd.read_csv("../Data/diabetes/labs.csv")
diet_df = pd.read_csv("../Data/diabetes/diet.csv")
examination_df = pd.read_csv("../Data/diabetes/examination.csv")
demographic_df = pd.read_csv("../Data/diabetes/demographic.csv")
questionnaire_df = pd.read_csv("../Data/diabetes/questionnaire.csv")
```

```python
diet_df.drop(columns='SEQN', inplace=True)
examination_df.drop(columns='SEQN', inplace=True)
demographic_df.drop(columns='SEQN', inplace=True)
questionnaire_df.drop(columns='SEQN', inplace=True)
```

```python
df = pd.concat([labs_df, diet_df], axis=1, join='inner')
df = pd.concat([df, examination_df], axis=1, join='inner')
df = pd.concat([df, demographic_df], axis=1, join='inner')
df = pd.concat([df, questionnaire_df], axis=1, join='inner')
df.head()
```

```python
df.info()
```

```python
df.dropna(axis=0, how='all', inplace=True)
df.dropna(axis=1, how='all', inplace=True)
```

```python
df.info()
```

According to Glycohemoglobin, 0-> Normal status 1-> High-risk diabetes 2-> Diabetes patients

```python
df = df.rename(columns={'SEQN' : 'ID',
                        'RIAGENDR' : 'Gender',
                        'DMDYRSUS' : 'Years_in_US',
                        'INDFMPIR' : 'Family_income',
                        'LBXGH' : 'GlycoHemoglobin',
                        'BMXARMC' : 'ArmCircum',
                        'BMDAVSAD' : 'SaggitalAbdominal',
                        'MGDCGSZ' : 'GripStrength',
                        'DRABF' : 'Breast_fed'})

df = df.loc[:, ['ID', 'Gender', 'Years_in_US', 'Family_income','GlycoHemoglobin', 'ArmCircum',  'SaggitalAbdominal', 
                'GripStrength', 'Breast_fed']]

df.describe()
```

```python
df.isnull()
```

```python
df['Years_in_US'] = df['Years_in_US'].apply(lambda x: x if x > 0 else 0)

df['GlycoHemoglobin'] = df['GlycoHemoglobin'].fillna(df['GlycoHemoglobin'].median())
df['SaggitalAbdominal'] = df['SaggitalAbdominal'].fillna(df['SaggitalAbdominal'].median())
df['ArmCircum'] = df['ArmCircum'].fillna(df['ArmCircum'].median())
df['GripStrength'] = df['GripStrength'].fillna(df['GripStrength'].median())

df['Family_income'] = df['Family_income'].fillna(method='ffill')

df['Breast_fed'] = df['Breast_fed'].fillna(value = 1)

df.describe()
```

```python
df.loc[df['GlycoHemoglobin'] < 6.0, 'Diabetes'] = 0
df.loc[(df['GlycoHemoglobin'] >= 6.0) & (df['GlycoHemoglobin'] <= 6.4), 'Diabetes'] = 1
df.loc[df['GlycoHemoglobin'] >= 6.5, 'Diabetes'] = 2

df.head(10)
```

```python
df.drop(['GlycoHemoglobin'], axis=1, inplace=True)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split

x = df.iloc[:, 4:].drop(['Diabetes'], axis=1)
y = df.iloc[:, 8].to_frame()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=500)
forest.fit(x_train, y_train)
res = forest.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, res)
```

0-> Normal status 1-> High-risk diabetes 2-> Diabetes patients

```python
from sklearn.ensemble import AdaBoostClassifier

ada_boost = AdaBoostClassifier(n_estimators=500)
ada_boost.fit(x_train, y_train)
res = ada_boost.predict(x_test)
```

```python
confusion_matrix(y_test, res)
```

```python
x = df.loc[:, ['Gender', 'Years_in_US', 'Family_income', 'GripStrength', 'Breast_fed']]
```

```python
x
```

```python
from sklearn.cluster import KMeans

k_means = KMeans(n_clusters=3)
k_means.fit_predict(x)
```

```python
plt.title("Before clustering")
plt.scatter(x.values[:, 2], x.values[:, 3])
plt.show()

plt.title("After clustering")
plt.scatter(x.values[:, 2], x.values[:, 3], c=k_means.labels_)
plt.show()
```

```python
for i in range(0, x.shape[0]):
    if k_means.labels_[i] == 1:
        c1 = plt.scatter(x.iloc[i, 2], x.iloc[i, 3], c='r')
    elif k_means.labels_[i] == 0:
        c2 = plt.scatter(x.iloc[i, 2], x.iloc[i, 3], c='g')
    elif k_means.labels_[i] == 2:
        c3 = plt.scatter(x.iloc[i, 2], x.iloc[i, 3], c='b')
        
plt.legend([c1, c2, c3],['High risk', 'Normal', 'Patient'])
plt.show()
```
